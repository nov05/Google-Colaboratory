# üü¢ **Local models**  

## üëâ **Ollama**

**make sure you've 1. installed ollama, downloaded (pulled) the model, and started "ollama serve" as explained in the book!**

```python
%%time
## Ollama, a platform for running LLMs locally
## Download its installer from https://ollama.com
## If you are using a CPU instance, it will take a few minutes.
from langchain_ollama import ChatOllama
chat = ChatOllama(
    model="deepseek-r1:1.5b",
    temperature=0,
)
messages = [
    ("system","You are a helpful assistant.",),
    ("human", "What makes LangChain great for working with LLMs?",),
]
ai_msg = chat.invoke(messages)
print(ai_msg.content)
## CPU times: user 1.51 s, sys: 120 ms, total: 1.63 s
## Wall time: 3min 44s
```

```python
from IPython.display import display, Markdown
display(Markdown(ai_msg.content))
```

## üëâ **Huggingface**  

```python
%%time
## ‚ö†Ô∏è Set up the HF_TOKEN secret in Colab and grand access.
from langchain_core.messages import SystemMessage, HumanMessage
from langchain_huggingface import ChatHuggingFace, HuggingFacePipeline
import os
os.environ["TRANSFORMERS_VERBOSITY"] = "info"
# Create a pipeline with a small model:
llm = HuggingFacePipeline.from_model_id(
    model_id="TinyLlama/TinyLlama-1.1B-Chat-v1.0",
    task="text-generation",
    pipeline_kwargs=dict(
        max_new_tokens=512,
        do_sample=False,
        repetition_penalty=1.03,
    ),
)
chat_model = ChatHuggingFace(llm=llm)
# Use it like any other LangChain LLM
messages = [
    SystemMessage(content="You're a helpful assistant"),
    HumanMessage(
        content="Explain the concept of machine learning in simple terms"
    ),
]
ai_msg = chat_model.invoke(messages)
print(ai_msg.content)
## CPU times: user 2min 25s, sys: 12.3 s, total: 2min 37s
## Wall time: 3min 14s
```
