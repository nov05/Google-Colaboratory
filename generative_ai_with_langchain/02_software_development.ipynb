## üëâ **google/codegemma-2b**  

````python
%%time
## ‚ö†Ô∏è Go to https://huggingface.co/google/codegemma-2b and acknowledge the license
## ‚ö†Ô∏è Enable the secret "HF_TOKEN" in Colab
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
checkpoint = "google/codegemma-2b"  ## Choose a more up-to-date model
## Load the model and tokenizer
model = AutoModelForCausalLM.from_pretrained(
    checkpoint,
    device_map="auto"
)
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
## CPU times: user 28.6 s, sys: 17.2 s, total: 45.9 s
## Wall time: 2min 24s
## System RAM 10.9 / 12.7 GB (if using a CPU instance)
````

````python
%%time
## The class `HuggingFacePipeline` was deprecated in LangChain 0.0.37 and will be removed in 1.0.
## langchain-huggingface package and should be used instead.
# from langchain.llms import HuggingFacePipeline
from langchain_huggingface import HuggingFacePipeline
# Create a text generation pipeline
pipe = pipeline(
    task="text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=500
)
# Integrate the pipeline with LangChain
llm = HuggingFacePipeline(pipeline=pipe)
# Define the input text
input_text = """
def calculate_primes(n):
    \"\"\"Create a list of consecutive integers from 2 up to N.

    For example:
    >>> calculate_primes(20)
    Output: [2, 3, 5, 7, 11, 13, 17, 19]
    \"\"\"
"""
# Use the LangChain LLM to generate text
output = llm(input_text)
## Device set to use cuda:0
## CPU times: user 14.7 s, sys: 0 ns, total: 14.7 s
## Wall time: 14.7 s
````

````python
%%time
from transformers import AutoModelForCausalLM, AutoTokenizer
# import torch
## https://huggingface.co/bigcode/starcoder2-3b
checkpoint = "bigcode/starcoder2-3b"  ## smaller model
model = AutoModelForCausalLM.from_pretrained(
    checkpoint,
    device_map="auto",
    # torch_dtype=torch.bfloat16,  ## lower precision
)
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
print(f"Memory footprint: {model.get_memory_footprint() / 1e6:.2f} MB")
## CPU times: user 28.5 s, sys: 22.5 s, total: 51 s
## Wall time: 2min 11s
## Memory footprint: 12121.49 MB
## Disk 46.2 / 112.6 GB
## GPU RAM 11.5 / 15.0 GB
````

````python
from transformers import pipeline
from langchain_huggingface import HuggingFacePipeline
pipe = pipeline(
    task="text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=2000,
    do_sample=True,
    temperature=0.5,
    top_p=0.95,
    repetition_penalty=1.1
)
llm = HuggingFacePipeline(pipeline=pipe)
## Device set to use cuda:0
````

````python
%%time
input_text = """
def calculate_primes(n):
    \"\"\"Create a list of consecutive integers from 2 up to N.

    For example:
    >>> calculate_primes(20)
    Output: [2, 3, 5, 7, 11, 13, 17, 19]
    \"\"\"
"""
output = llm.invoke(input_text)
## Setting `pad_token_id` to `eos_token_id`:0 for open-end generation.
## CPU times: user 27.2 s, sys: 313 ms, total: 27.5 s
## Wall time: 28.6 s
````

````python
print(output)
````

## üëâ **claude-3-opus-20240229**

````python
%%time
from langchain_anthropic import ChatAnthropic
from langchain_core.prompts.prompt import PromptTemplate
from google.colab import userdata
input_text = """
def calculate_primes(n):
    \"\"\"Create a list of consecutive integers from 2 up to N.

    For example:
    >>> calculate_primes(20)
    Output: [2, 3, 5, 7, 11, 13, 17, 19]
    \"\"\"
"""
template = """Question: {question}
Let's think step by step.

Answer:
"""
prompt = PromptTemplate(template=template, input_variables=["question"])
## Get API key at https://console.anthropic.com/settings/keys
llm = ChatAnthropic(
    model='claude-3-opus-20240229',
    api_key=userdata.get('ANTHROPIC_API_KEY')  ## ‚úÖ
)
llm_chain = prompt | llm
output = llm_chain.invoke(input_text)
## CPU times: user 325 ms, sys: 30.8 ms, total: 356 ms
## Wall time: 24.4 s
## Cost $0.11
````

## üëâ **LangChain agent with fake list LLM**  

````python
from langchain_core.language_models.fake import FakeListLLM
from langchain.agents import initialize_agent, AgentType
from langchain_experimental.tools import PythonREPLTool
responses = ["Action: Python_REPL\nAction Input: print(2 + 2)", "Final Answer: 4"]
llm = FakeListLLM(responses=responses)
tools = [PythonREPLTool()]
agent = initialize_agent(
    tools,
    llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
)
result = agent("What is 2 + 2?")
## CPU times: user 1.4 s, sys: 228 ms, total: 1.63 s
## Wall time: 1.98 s
````

````python
result
````

## üëâ **LangGraph ReAct agent with OpenAI chat model**  

````python
from langchain_openai import ChatOpenAI
from langgraph.prebuilt import create_react_agent
from langchain_experimental.tools import PythonREPLTool
from google.colab import userdata
from langchain_core.messages import HumanMessage
llm = ChatOpenAI(
    model="gpt-4o-mini",
    api_key=userdata.get('OPENAI_API_KEY'),
)
tools = [PythonREPLTool()]
agent = create_react_agent(llm, tools)
result = agent.invoke({"messages": [HumanMessage(content="What are the prime numbers until 20?")]})
````
