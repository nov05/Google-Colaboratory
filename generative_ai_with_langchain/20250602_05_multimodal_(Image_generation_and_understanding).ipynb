### **✅ With Stability AI**  

```python
## https://platform.stability.ai/docs
import requests
import os
from google.colab import userdata # Ensure userdata is imported

os.environ['STABILITY_API_KEY'] = userdata.get('STABILITY_API_KEY')  ## ✅
url = "https://api.stability.ai/v2beta/stable-image/generate/core"
headers = {
    "Authorization": f"Bearer {os.environ['STABILITY_API_KEY']}",
    "Accept": "application/json",
    # "Content-Type": "multipart/form-data",  ## ❌ do not add this line
}
files = {
    "mode": (None, "text-to-image"),
    "prompt": (None, "A detailed technical diagram of an AI agent"),
    "aspect_ratio": (None, "1:1"),
    "output_format": (None, "webp"),
    "cfg_scale": (None, "4.5"),  # <- as string
    "steps": (None, "40")        # <- as string
}
response = requests.post(url, headers=headers, files=files)

# Print the response content before raising an error
print("Response Status Code:", response.status_code)
try:
    print("Response Body:", response.json())
except requests.exceptions.JSONDecodeError:
    print("Response Body (non-JSON):", response.text)
response.raise_for_status()  # Will raise HTTPError if 4xx or 5xx
result = response.json()
print(result)
```

```python
import base64
image_base64 = response.json()['image']
image_bytes = base64.b64decode(image_base64)
with open("image.png", "wb") as f:
    f.write(image_bytes)
print("Image saved as image.png")
```

## Image understanding

```python
from langchain_core.messages import HumanMessage
from langchain_openai import ChatOpenAI

def analyze_image(image_url: str, question: str) -> str:
    chat = ChatOpenAI(model="gpt-4o-mini", max_tokens=256)
    message = HumanMessage(
        content=[
            {
                "type": "text",
                "text": question
            },
            {
                "type": "image_url",
                "image_url": {
                    "url": image_url,
                    "detail": "auto"
                }
            }
        ]
    )
    response = chat.invoke([message])
    return response.content

# Example usage
# image_url = "https://replicate.delivery/yhqm/JGTpV46FdT7dE5RGgs0dy5qgLegwOqoIPPaAvrfOmOGMwkMUA/out-0.png"
imge_url = "https://raw.githubusercontent.com/nov05/pictures/refs/heads/master/pic001/20250602_dalle_img-NcnHK9okkXjdTzGfFTCoPcxl.png"
questions = [
    "What objects do you see in this image?",
    "What is the overall mood or atmosphere?",
    "Are there any people in the image?"
]
for question in questions:
    print(f"\nQ: {question}")
    print(f"A: {analyze_image(image_url, question)}")
```

```python
import google.generativeai as genai
genai.configure(api_key=os.environ['GOOGLE_API_KEY'])
video_file_name = "ForBiggerJoyrides.mp4"
print(f"Uploading file...")
video_file = genai.upload_file(path=video_file_name)
print(f"Completed upload: {video_file.uri}")
```

```python
%%time
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.messages import HumanMessage
prompt = [
    {
        "type": "text",
        "text": "Describe the video in a few sentences."
    },
    {
        "type": "media",
        "file_uri": video_file.uri,
        "mime_type": "video/mp4"
    },
]
llm = ChatGoogleGenerativeAI(
    model="gemini-1.5-pro",
    temperature=0,
)
response = llm.invoke([HumanMessage(content=prompt)])
print(response.content)
```
